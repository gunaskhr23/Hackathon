```python
import pandas as pd
df = pd.read_csv('final_food_delivery_dataset.csv')
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True)
```

```python
gold_members = df[df['membership'] == 'Gold']
city_gold_rev = gold_members.groupby('city')['total_amount'].sum().sort_values(ascending=False)
print("City with highest Gold member revenue:")
```

```python
gold_members = df[df['membership'] == 'Gold']
city_gold_rev = gold_members.groupby('city')['total_amount'].sum().sort_values(ascending=False)
print("City with highest Gold member revenue:")
print(city_gold_rev.head(1))
```

```python
cuisine_aov = df.groupby('cuisine')['total_amount'].mean().sort_values(ascending=False)
print("Cuisine with highest Average Order Value:")
print(cuisine_aov.head(1))
```

```python
user_total_spend = df.groupby('user_id')['total_amount'].sum()
high_spenders = user_total_spend[user_total_spend > 1000].count()
print(f" Number of distinct users  : {high_spenders}")
```

```python
bins = [3.0, 3.5, 4.0, 4.5, 5.0]
df['rating_range'] = pd.cut(df['rating'], bins=bins)
rating_rev = df.groupby('rating_range', observed=False)['total_amount'].sum().sort_values(ascending=False)
print(" Rating range with highest revenue:")
print(rating_rev.head(1))
```

```python
gold_city_aov = gold_members.groupby('city')['total_amount'].mean().sort_values(ascending=False)
print("City with highest AOV among Gold members:")
print(gold_city_aov.head(1))
```

```python
cuisine_stats = df.groupby('cuisine').agg({'restaurant_name_x': 'nunique', 'total_amount': 'sum'})
cuisine_stats.columns = ['distinct_restaurants', 'total_revenue']
cuisine_stats_sorted = cuisine_stats.sort_values('distinct_restaurants')
print(f"Cuisine with lowest distinct restaurants but significant revenue: {cuisine_stats_sorted.index[0]}")
```
